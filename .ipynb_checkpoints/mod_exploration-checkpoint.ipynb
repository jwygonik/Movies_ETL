```python
# Make an empty list to hold all the alt titles

# def clean_movie(movie):
    # movie = dict(movie) # create non-destructive copy
    # alt_titles={}
    # return movie
```

```python
# Loop through list of all alt keys

def clean_movie(movie):
    movie = dict(movie) 
    alt_titles = {}
    for key in ['Also known as','Arabic','Cantonese','Chinese','French',
                'Hangul','Hebrew','Hepburn','Japanese','Literally',
                'Mandarin','McCune–Reischauer','Original title','Polish',
                'Revised Romanization','Romanized','Russian',
                'Simplified','Traditional','Yiddish']:
        
        if key in movie:  # check to see if the current key exist in the movie obj
            alt_titles[key] = movie[key] # if exists, remove key-value pair & add to alt title dic
            movie.pop(key)
        if len(alt_titles) > 0:
            movie['alt_titles'] = alt_titles    # after looping thru every key, add the alt titles to dict
            
    return movie
```

```python
# make a list of cleaned movies with list comprehension

clean_movies = [clean_movie(movie) for movie in wiki_movies]
```

```python
# set wiki_movies_df to be the DF created from clean_movies and print out list of columns

wiki_movies_df = pd.DataFrame(clean_movies)
sorted(wiki_movies_df.columns.tolist())
```
